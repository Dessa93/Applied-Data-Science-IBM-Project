```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
from folium.features import DivIcon

# Coordinates
launch_site_lat = 28.562302
launch_site_lon = -80.577356
coastline_lat = 28.56367
coastline_lon = -80.57163

# Calculate distance
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Create a marker on the coastline point
distance_marker = folium.Marker(
    location=[coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_coastline),
    )
)

# Add marker to the map
distance_marker.add_to(site_map)

# Optionally, add a line between launch site and coastline for visualization
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]],
    color='blue',
    weight=2,
    dash_array='5,5'
).add_to(site_map)

# Display map
site_map
```

```python
from folium.features import DivIcon
import folium

# Example launch sites and their nearest points of interest
# Replace POIs with closest city, highway, railway, etc.
launch_sites = {
    "CCAFS LC-40": {"coord": [28.562302, -80.577356], "poi": [28.56367, -80.57163]},
    "CCAFS SLC-40": {"coord": [28.563197, -80.576820], "poi": [28.56500, -80.57000]},
    "KSC LC-39A": {"coord": [28.573255, -80.646895], "poi": [28.57500, -80.64000]},
    "VAFB SLC-4E": {"coord": [34.632834, -120.610745], "poi": [34.63500, -120.60500]}
}

# Initialize map
site_map = folium.Map(location=[30, -90], zoom_start=5)

# Loop through each launch site
for site, data in launch_sites.items():
    launch_coord = data["coord"]
    poi_coord = data["poi"]
    
    # Calculate distance
    distance = calculate_distance(launch_coord[0], launch_coord[1],
                                  poi_coord[0], poi_coord[1])
    
    # Marker at the POI
    poi_marker = folium.Marker(
        location=poi_coord,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance)
        )
    )
    poi_marker.add_to(site_map)
    
    # Line from launch site to POI
    line = folium.PolyLine(
        locations=[launch_coord, poi_coord],
        color='green',
        weight=2,
        dash_array='5,5'
    )
    site_map.add_child(line)
    
    # Optional: Add a marker for the launch site itself
    folium.Circle(
        location=launch_coord,
        radius=500,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=site
    ).add_to(site_map)

# Display map
site_map
```

1. Proximity to Railways

Most launch sites, especially CCAFS, KSC, and VAFB, are very close to railways (within a few kilometers).

Why: Railways are used to transport large rocket components (like Falcon 9 cores or payloads) from manufacturing facilities to the launch site.

2. Proximity to Highways

All launch sites are very close to major highways.

Reason: Highways allow the movement of heavy equipment and provide access for personnel, making logistics and emergency access easier.

3. Proximity to Coastline

All sites are directly on or extremely close to the coast:

Florida sites  Atlantic coast

Vandenberg  Pacific coast

Why: Rockets are launched over the ocean for safety (minimizing risk to populated areas) and to allow multiple orbital trajectories.

4. Distance from Cities

Launch sites are intentionally located several kilometers away from populated areas.

Reason: In case of a launch failure, debris or explosion risk to humans and property is minimized.

Example: CCAFS and KSC are in sparsely populated areas of Florida; VAFB is in a remote region of California.

 Findings Summary
Feature	Proximity	Purpose
Railways	Close	Transport rocket parts
Highways	Close	Logistics and access
Coastline	Very close	Safe over-ocean launches
Cities	Far enough	Safety from debris/explosions
